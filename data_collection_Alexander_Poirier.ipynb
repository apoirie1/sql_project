Start of building data for trips in our database
    will use trail and user id, 
    randomly generate numbers for review and date

```python
# list of trail IDs
t_id = trail_data.get('Id')
```

```python
type(t_id)
```

```python
# make new dictionary for trips table
import random
import datetime

trip_data = {
    'tripID':[],
    'userID':[],
    'trailID':[],
    'date':[],
    'review':[],
}
    # Trip ID 
    # User ID 
    # Trail ID
    # Date 
    # Review (1-5 with one decimal place)
trip_id = 0    
    
for i in range (1000):
    trip_data['tripID'].append(trip_id)
    
    # using users 1-60 since we dont have more users yet
    trip_data['userID'].append(random.randint(41,81))
    
    random_trail = random.choice(t_id)
    trip_data['trailID'].append(random_trail)
    
    start = datetime.date(2020, 5, 10)
    end = datetime.date(2020, 12, 31)
    time_between_dates = end - start
    days_between_dates = time_between_dates.days
    random_number_of_days = random.randrange(days_between_dates)
    random_date = start + datetime.timedelta(days=random_number_of_days)
    trip_data['date'].append(random_date)
    
    random_review = random.randint(0,4)+(random.randint(0,9)/10)
    trip_data['review'].append(random_review)
    
    
    trip_id = trip_id + 1
    
    
```

```python
trip_data
```

```python
trip_df = pd.DataFrame(trip_data)
```

```python
trip_df.head()
```

```python
trip_df.rename({'tripID': 'trip_id', 'userID': 'user_id', 'trailID': 'trail_id'}, axis=1, inplace=True)
```

```python
trip_df.drop_duplicates(subset = 'trip_id', inplace = True)
```

```python
# add to sql 
trip_df.to_sql('trips', engine, if_exists='append',index=False)
```
