Main Business Question: When looking at the summertime, how many hikes were taken by users? Compare these resuults to the fall.

Business Justification: Knowing that summer is a popular time for individuals to be outdoors and active, we as a company will want to look durinig peak season and see how many hikes we have logged. This will help us to see if as a company we are succeeding during a season that we expect volume to be at its highest.

Features: VIEW, JOIN, CTE

```python
%%sql
CREATE VIEW summer_hikes AS
SELECT 
    trip_id,
    trips.date,
    trips.user_id,
    full_name,
    trips.trail_id,
    trail_name,
    location,
    trails.length
FROM trips
JOIN users
    ON users.user_id = trips.user_id
JOIN trails 
    ON trails.trail_id = trips.trail_id
WHERE 
    trips.date BETWEEN '2020-05-10' AND '2020-08-10';
```

```python
%%sql
SELECT *
FROM summer_hikes;
```

Analysis: There were 392 hikes that were logged for our users between May 10 and August 10 of 2020.

SQL below to compare summer to the fall hiking to see if there is a difference

```python
%%sql
WITH fallHikesCTE AS (
   SELECT 
        trip_id,
        trips.date,
        trips.user_id,
        full_name,
        trips.trail_id,
        trail_name,
        location,
        trails.length
    FROM trips
    JOIN users
        ON users.user_id = trips.user_id
    JOIN trails 
        ON trails.trail_id = trips.trail_id
    WHERE 
        trips.date BETWEEN '2020-08-15' AND '2020-11-15'
)
SELECT *
FROM fallHikesCTE;
```

When comparing the two segments, we see almost an identical number of hikes occuring in both the summer and fall. This can be explained due to thee consistency of weather in Calfornia. Other states often have larger ranges in terms or weather and temperature where the users here on the west coast are used to sunny and warm days year round, especially in the south. Since this is a set of mock data, it makes sense to see this since thee date was just a random variable, but we may  be surprised to see similar real life results to this due to the weather and active culture found on the west coast. There is a lot of good hiking year round in California, and we can expect our users to take advantage of it.

Business Query #2 - Looking in summer_hikes, lets rank based on location to see if we can find some popular trails

Business Justification: this kind of query will help Inspire Outdoors to locate areas of popularity. We can see where our users hike, how many trips are being logged there (regardless of the individual) and hopefully we can continue to see growth in all trails withmore users signinig up and more hikes being logged. 

Features: WINDOW, GROUP BY

```python
%%sql
SELECT
    trip_id,
    trail_name,
    location,
    COUNT(trip_id) AS number_of_trips,
    DENSE_RANK() OVER (
        PARTITION BY location
        ORDER BY COUNT(trip_id) DESC
    ) AS trip_count_location_rank
FROM summer_hikes
GROUP BY trail_id;
```

WE can see majoirty of trails were only hiked once and can be in a tie for the most hiked given that specific location. Thiis is due to the limited number of entries we have for our trips table. Hoopefully with more data, we will be able to see larger numbers to signify enjoyment of our app from all different locations. One important note is we see that Three Rivers is a very popular area, and has over 10 trails that are in the top 2 most hiked. We can focus on Three Rivers in the next query to gain insight into how users feel about the area.

Business Query #3 - We found in the previous Query that Three Rivers California is a popular spot to hike with over 20 hikes occuring there this summer, lets take a look at all of the hikes that were logged at Three Rivers, and seeing how their ratings are. This query will return the trail name, a rating, the user who went on the trip, and finally an additional piece added with a CASE statment to determine if the review was bad, ok, or great.

Business Justification: this query helps us to see how users are responding to this popular area. We are hoping to see positive reviews, as this indicates customer satisfaction and will likely lead to them returning to the app in order to add new hikes in the future. If we see negative reviews, we can hoopefully reach out to these users to see if we can figure out improvements

Features: CASE, Subquery

```python
%%sql
 SELECT
   	users.full_name,
    trails.trail_name,
    location, 
    trips.review,
    CASE
        WHEN trips.review = 0 OR trips.review = 1 THEN 'Poor'
        WHEN trips.review = 2 OR trips.review = 3 THEN 'Good'
        WHEN trips.review = 4 OR trips.review = 5 THEN 'Excellent'
    END AS user_experience
FROM trips
JOIN users
    ON users.user_id = trips.user_id
JOIN trails 
    ON trails.trail_id = trips.trail_id
WHERE location IN (
	SELECT 
		location
	FROM trails 
	WHERE location LIKE '%Three Rivers%'
)
ORDER BY user_experience;
```

Analysis: When looking at the results of the querey, it seems that most of the trips that were logged at Three Rivers were pretty positive. There were a handful of Poor and Excellent but majoirty of people rated their adventure in our Good category. This lets us see that users enjoyed their time at Three Rivers and we can hopefully assume that they will return to the area.
